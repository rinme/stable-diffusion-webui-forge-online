# Download the models, using civitai cookies (Because I didn't know they had api token thing)

```python
!apt-get update
!apt-get install aria2 -y
%cd /notebooks/

import os

def download(download_url, output_location, cookie="", output_name=""):
    """Downloads a file using aria2c, optionally providing a cookie and specifying the output location.

    Args:
        download_url (str): The URL of the file to download.
        output_location (str): The desired output directory or full path for the downloaded file.
        cookie (str, optional): The cookie string to include in the request. Defaults to "".
    """

    aria2c_command = f'aria2c -s16 -x16 --continue=true -d "{output_location}" "{download_url}"'

    # Add the cookie header if a cookie is provided
    if cookie:
        aria2c_command = f'aria2c -s16 -x16 --continue=true --header="Cookie:__Secure-civitai-token={cookie}" -d "{output_location}" "{download_url}"'

    if cookie and output_name:
        aria2c_command = f'aria2c -s16 -x16 --continue=true --header="Cookie:__Secure-civitai-token={cookie}" -d "{output_location}" -o "{output_name}" "{download_url}"'

    os.system(aria2c_command) 

# Example usage with cookie:
civitai_cookie = "" 


# Checkpoints

# SD1.5
download("https://huggingface.co/runwayml/stable-diffusion-v1-5/resolve/main/v1-5-pruned-emaonly.ckpt", "./stable-diffusion-webui-forge-online/models/Stable-diffusion/")
## Pony
download("https://civitai.com/api/download/models/290640?type=Model&format=SafeTensor&size=pruned&fp=fp16", "./stable-diffusion-webui-forge-online/models/Stable-diffusion/", civitai_cookie, "ponyDiffusionV6XL_v6StartWithThisOne.safetensors")
## Forget
download("https://civitai.com/api/download/models/324619", "./stable-diffusion-webui-forge-online/models/Stable-diffusion/", civitai_cookie, "autismmixSDXL_autismmixPony.safetensors")
## Anything Mix XL
download("https://civitai.com/api/download/models/384264?type=Model&format=SafeTensor&size=full&fp=fp16", "./stable-diffusion-webui-forge-online/models/Stable-diffusion/", civitai_cookie, "AnythingXL_xl.safetensors")


# VAE
download("https://civitai.com/api/download/models/290640?type=VAE&format=SafeTensor", "./stable-diffusion-webui-forge-online/models/VAE/", civitai_cookie)


# Loras
## Hazel
download("https://civitai.com/api/download/models/420888?type=Model&format=SafeTensor", "./stable-diffusion-webui-forge-online/models/Lora/", civitai_cookie, "Hazel_XL_Pony.safetensors")
download("https://civitai.com/api/download/models/333607?type=Model&format=SafeTensor", "./stable-diffusion-webui-forge-online/models/Lora/", civitai_cookie, "Smooth Anime Style LoRA XL.safetensors")
download("https://civitai.com/api/download/models/401789?type=Model&format=SafeTensor", "./stable-diffusion-webui-forge-online/models/Lora/", civitai_cookie, "merunyaa_style_ponyxl_v1_unet.safetensors")


# Embeddings
download("https://civitai.com/api/download/models/60095?type=Negative&format=Other", "./stable-diffusion-webui-forge-online/embeddings/", civitai_cookie, "bad_prompt_version2.pt")
```
